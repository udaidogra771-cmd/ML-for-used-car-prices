*italicized text*### 1. VARIABLES: THE STORAGE BOXES

In Scratch: [set car_brand to "Ford"]

In Python: Use the '=' sign to put something inside a "box".

```python
car_brand = "Ford"
price = 15000
is_electric = False
```

```python
# Printing (like the 'Say' block in Scratch)
print("--- 1. VARIABLES ---")
print("Car Brand:", car_brand)
print("Price tag:", price)
print("Is it electric?", is_electric)
print()
```

### 2. LISTS: THE PARKING LOT

In Scratch: You have Lists.

In Python, we use square brackets [ ].

**IMPORTANT:** In Python, we start counting at 0!

```python
garage = ["Mustang", "Camaro", "F-150", "Corvette"]

print("--- 2. LISTS ---")
print("My full garage:", garage)
# Scratch: [item 1 of garage] -> Python: garage[0]
print("The first car in the lot is:", garage[2])
print("")
```
